```python
import cv2
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#read image
my_img=cv2.imread('./flower.jpg',cv2.IMREAD_GRAYSCALE)
my_img.ndim
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(my_img)
```

```python
#show image with plt
plt.imshow(my_img,'gray')
```

```python
#cvtColor : Convert image color
my_img_rgb=cv2.cvtColor(my_img,cv2.COLOR_GRAY2RGB)
print(my_img_rgb.ndim)
```

```python
# rgb image vs bgr
my_img_bgr=cv2.imread('./flower.jpg',cv2.IMREAD_COLOR) #always cv2 read in BGR Color
my_img_rgb=cv2.cvtColor(my_img_bgr,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(my_img_bgr);plt.title("BGR Image");plt.show()
plt.imshow(my_img_rgb);plt.title("RGB Image");plt.show()
```

```python
for i in range(my_img_bgr.shape[2]):
  plt.subplot(1,3,i+1)
  plt.imshow(my_img_bgr[...,i])
plt.show()
```

```python
my_img_yrb=cv2.cvtColor(my_img_rgb,cv2.COLOR_RGB2YCrCb)
plt.imshow(my_img_yrb)
```

```python
for i in range(my_img_yrb.shape[2]):
  plt.subplot(1,3,i+1)
  plt.imshow(my_img_yrb[...,i],cmap='gray')
plt.show()
```
